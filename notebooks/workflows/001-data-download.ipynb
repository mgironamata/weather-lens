```python
from herbie import Herbie
import os
from datetime import datetime, timedelta
```

```python
def tp_download(DATE, fxx=24, dest_dir=None):
    # RETRIEVE
    HAIFS = Herbie(DATE, model="aifs", product="enfo", fxx=fxx, param='TP')
    HIFS = Herbie(DATE, model="ifs", product="enfo", fxx=fxx, param='TP')
    
    HAIFS_NAME = f"AIFS_{DATE}_TP_FXX24.grib2"
    HIFS_NAME = f"IFS_{DATE}_TP_FXX24.grib2"
    
    HAIFS.LOCALFILE = HAIFS_NAME
    HIFS.LOCALFILE = HIFS_NAME
    
    # DOWNLOAD
    PATH_AIFS = HAIFS.download(search=":tp:",verbose=True, overwrite=True, save_dir=dest_dir)
    PATH_IFS = HIFS.download(search=":tp:",verbose=True, overwrite=True, save_dir=dest_dir)

    # Define the source and destination paths
    # source_aifs = PATH_AIFS
    # source_ifs = PATH_IFS
    # destination_dir = dest_dir
    
    # if destination_dir is not None:
    
    #     # Create the destination directory if it doesn't exist
    #     os.makedirs(destination_dir, exist_ok=True)
        
    #     # Move the files
    #     !mv "{source_aifs}" "{destination_dir}"
    #     !mv "{source_ifs}" "{destination_dir}"
        
    #     print(f"Moved {source_aifs} to {destination_dir}")
    #     print(f"Moved {source_ifs} to {destination_dir}")
```

```python
variable = '2t' #e.g., tp, cape, etc.
fxx = 24  #forecast hours, e.g., 24, 48, etc.
dest_path = f"../aifs-data/{variable}/{fxx}/"
```

```python
# start_date = datetime(2025, 7, 2)
start_date = datetime(2025, 9, 7)
date_list = [start_date + timedelta(days=i) for i in range(90)]
formatted_dates = [date.strftime("%Y-%m-%d-06") for date in date_list]

for date_str in formatted_dates:
  # stop if date is > 29 Sept 2025
  if datetime.strptime(date_str, "%Y-%m-%d-%H") > datetime(2025, 9, 29, 23):
      break
  print("Starting with date:", date_str)
  tp_download(date_str, dest_dir=dest_path, fxx=fxx)
```

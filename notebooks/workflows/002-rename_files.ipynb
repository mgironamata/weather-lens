```python
from pathlib import Path
import re

# set your target directory
dir_path = Path("/scratch2/aifs-data/tp/")

# regex for leading "subset_<hash>__"
pattern = re.compile(r"^subset_[a-z0-9]+__")

for f in dir_path.iterdir():
    if f.is_file():
        new_name = pattern.sub("", f.name)
        if new_name != f.name:
            new_path = f.with_name(new_name)
            print(f"Renaming: {f.name} -> {new_name}")
            f.rename(new_path)
```
